# Task 1: Exploring and Visualizing the Iris Dataset
**Intern:** Dua Ilyas
**Objective:** To understand data structure, summarize key statistics, and visualize relationships between features of the Iris flower.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from seaborn's built-in library
df = sns.load_dataset('iris')

# Display basic structure
print("1. DATASET SHAPE:", df.shape)
print("\n2. COLUMN NAMES:", df.columns.tolist())
print("\n3. FIRST 5 ROWS:")
print(df.head())
```

```python
# Check for missing values (Required for good data practice)
print("Missing values in each column:\n", df.isnull().sum())

# Statistical summary
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Set the visual style
sns.set_theme(style="whitegrid")

# 1. Scatter Plot (Relationship between Sepal Length and Width)
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', s=70)
plt.title('Sepal Length vs. Sepal Width')
plt.show()

# 2. Histogram (Distribution of Petal Length)
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='petal_length', kde=True, color='teal')
plt.title('Distribution of Petal Length')
plt.show()

# 3. Box Plot (To detect outliers and spread)
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='species', y='petal_width')
plt.title('Petal Width Distribution by Species')
plt.show()
```

## Conclusion and Insights
1. **Species Separation:** In the scatter plot, the *Setosa* species is clearly separated from *Versicolor* and *Virginica*.
2. **Data Distribution:** The histogram shows that petal lengths are not normally distributed; there is a clear gap between smaller and larger petals.
3. **Outliers:** The box plot helps identify that some species have a wider range of petal widths than others, with very few outliers.
